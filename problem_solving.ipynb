```python
def max_profit(prices):
    if len(prices) < 2:
        return 0

    min_num  = prices[0]
    profit = 0
    """
    catch the min number & update it then check 
    if iterable num - min >  profit
    update the profit  
    """   
    for i in range(1, len(prices)):
        if prices[i] < min_num :
            min_num  = prices[i]
        elif prices[i] - min_num  > profit:
            profit = prices[i] - min_num 

    return profit

# max_profit([7, 1, 5, 3, 6, 4])
# max_profit([7,6,4,3,1])
# max_profit([2,1,4])
max_profit([2,4,1])
```

```python
roman_dict = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}
roman_str = "IIIV"

total_sum = 0
last_num = 0

for i in range(len(roman_str)):
    current_num = roman_dict[roman_str[i]]  # 1
    if current_num > last_num:
        total_sum += current_num - 2 * last_num  # 1   3 + (5-2) 
    else:
        total_sum += current_num  # 1+1+1
    last_num = current_num 

print(total_sum)
```

```python
def generateParentheses(n):
    result = []
    generate("", n, n, result)
    return result

def generate(current, left, right, result):
    if left == 0 and right == 0:
        result.append(current)
        return
    
    if left > 0:
        generate(current + "(", left - 1, right, result)
    
    if left < right:
        generate(current + ")", left, right - 1, result)

# Example usage
n = 7
combinations = generateParentheses(n)
print(combinations)
```
